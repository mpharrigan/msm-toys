```python
def rebuild(nsteps):
    t1 = toy.sample(0, nsteps)
    t1_counts = msml.get_count_matrix_from_assignments(t1[np.newaxis, :], lag_time=1, n_states=toy.n_states)
    t1_revcounts, t1_tmat, _, _ = msml.build_msm(t1_counts, symmetrize='MLE', ergodic_trimming=True)
    if t1_revcounts.shape[0] != toy.n_states:
        t1_revcounts, t1_tmat, _, _ = msml.build_msm(t1_counts, symmetrize='Transpose', ergodic_trimming=False)
    return np.linalg.norm(toy.tmat.toarray() - t1_tmat.toarray(), ord='fro')
def plot_vs(oom_range):
    xpoints = [int(10**oo) for oo in oom_range]
    edat = np.zeros((len(xpoints), 2))
    for i, nstep in enumerate(xpoints):
        edat[i, 0] = nstep
        edat[i, 1] = rebuild(nstep)
    return edat
```

```python
msml.logger.setLevel(logging.CRITICAL)
```

```python
t1_edat = plot_vs(np.linspace(1,5, 30))
fig, ax = plt.subplots()
ax.plot(t1_edat[:,0], t1_edat[:,1], 'o-')
ax.set_xscale('log'); ax.set_yscale('log')
```
